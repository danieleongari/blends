```python
import sys
sys.path.append('..')

import json

from blends.base import dict_to_blend

from blends.sample import get_samples_both, add_prop_to_samples


myblend_dict = json.load(open('../data/blend_example_010.json'))
myblend = dict_to_blend(myblend_dict)
df_samples = get_samples_both(
    myblend, 
    nsamples_vert=1000,
    nsamples_fill=30000, 
    qmethod="relative",
    verbose=True
    )
df_samples_price = add_prop_to_samples(myblend, df_samples, prop_name="price")
```

```python
from blends.viz import plot_design_space

for method in ["pca-2d", "pca-3d", "umap-2d", "umap-3d", "tsne-2d", "tsne-3d", ]:
    fig = plot_design_space(
            rblend=myblend,
            df_samples=df_samples_price,
            color="price",
            method=method,
            verbose=True,
            )
    fig.update_layout(width=700, height=600)
    fig.show()
```
